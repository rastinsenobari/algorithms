```python
goal = "123456780"
l = 3
c = 3
```

```python
class Node:
    global goal
    global l
    global c 

    def __init__(self, value, parrent):
        self.value = value
        self.parrent = parrent
        self.gscore = 0 if parrent==None else parrent.gscore + 1
        self.hscore = self.h()
        self.fscore = self.gscore + self.hscore 

    def find_index(self, puzzele, val):
        index = puzzele.index(str(val))

        return index//c, index%c

    def h(self):
        cost = 0
        for n in self.value:
            px, py = self.find_index(self.value, n)
            gx, gy = self.find_index(goal, n)
            cost += abs(gx-px) + abs(py-gy) 
        return cost
    
    def get_childs(self):
        childs = list()        

        x, y = self.find_index(self.value, 0)
        if x < l-1:
            tmp=list(self.value)
            tmp[x*c+y], tmp[(x+1)*c+y] = tmp[(x+1)*c+y], tmp[x*c+y]

            child = Node("".join(tmp), self)
            if child.parrent and child.parrent.parrent:
                if child.value != child.parrent.parrent.value:
                    childs.append(child)
            else:
                childs.append(child)
                
        if x > 0:
            tmp=list(self.value)
            tmp[x*c+y], tmp[(x-1)*c+y] = tmp[(x-1)*c+y], tmp[x*c+y]

            child = Node("".join(tmp), self)
            if child.parrent and child.parrent.parrent:
                if child.value != child.parrent.parrent.value:
                    childs.append(child)
            else:
                childs.append(child)
                
        if y < c-1:
            tmp=list(self.value)
            tmp[x*c+y], tmp[x*c+y+1] = tmp[x*c+y+1], tmp[x*c+y]

            child = Node("".join(tmp), self)
            if child.parrent and child.parrent.parrent:
                if child.value != child.parrent.parrent.value:
                    childs.append(child)
            else:
                childs.append(child)
                
        if y > 0:
            tmp=list(self.value)
            tmp[x*c+y], tmp[x*c+y-1] = tmp[x*c+y-1], tmp[x*c+y]
            
            child = Node("".join(tmp), self)
            if child.parrent and child.parrent.parrent:
                if child.value != child.parrent.parrent.value:
                    childs.append(child)
            else:
                childs.append(child)
                
        return childs
```

```python
def draw(current):
    global l
    global c
    step=current.gscore
    while current:
        for i in range(l):
            for j in range(c):
                print(current.value[i*3+j], end='')
            print()
        print()
        current = current.parrent
    print(f"we solve it in {step} step")
    return True
```

```python
def A_star(start, goal):
    open_list = [start]

    while len(open_list):
        current = open_list.pop(0)

        if current.value == goal:
            return draw(current)

        for child in current.get_childs():
            for i in range(len(open_list)):
                if child.value == open_list[i].value:
                    if child.fscore < open_list[i].fscore:
                        open_list[i] = child
                    break
                if child.fscore < open_list[i].fscore:
                    open_list.insert(i, child)
                    break
            else:
                open_list.append(child)
           
    return False
```
